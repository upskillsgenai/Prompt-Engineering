## **Chain of Thought**

```python
# System message with CoT instructions
system_message_cot = """I am a Legal Advisor Bot. I provide general legal information (not personalized advice).
Always consult a real attorney for legal matters.

### Chain-of-Thought Rules:
For every question, break your answer into:
1. Legal Concept: Identify the relevant law/principle.
2. Steps: List actionable steps for the user.
3. Disclaimer: Remind to consult a lawyer.
"""
```

```python
def get_legal_response_cot(user_query):
    response = client.chat.completions.create(
        model=os.getenv("AZURE_OAI_DEPLOYMENT", "gpt-4o-mini"),
        temperature=0.5,  # Moderate for balanced reasoning
        messages=[
            {"role": "system", "content": system_message_cot},
            {"role": "user", "content": f"Use Chain-of-Thought reasoning. Question: {user_query}"}
        ],
    )
    return response.choices[0].message.content
```

```python
# Test your query

print(get_legal_response_cot("How do I evict a tenant?"))
```

# Role-Play Prompting with Contextual Data (.txt)

```python
# Load context from .txt file
with open("compliance_data.txt", "r", encoding="utf-8") as f:
    context_data = f.read()

# Role-specific system prompt
system_message_roleplay = f"""
You are acting as a Senior Legal Compliance Officer.
Your job is to review questions and answer **based strictly on the provided internal compliance documentation**.

ğŸ“„ Reference Material:
\"\"\"
{context_data}
\"\"\"

ğŸ“ Guidelines:
- Only respond using information from the reference above.
- If the data is insufficient, politely say: "This information is not covered in the current compliance documentation."
- Use bullet points for steps or checklists.
- Keep tone professional and concise.
"""

# Prompt user for question
user_query = input("Enter your legal/compliance question: ")

# Get response using Azure OpenAI
def get_response_from_context(query):
    response = client.chat.completions.create(
        model=azure_oai_deployment,
        temperature=0.3,
        messages=[
            {"role": "system", "content": system_message_roleplay},
            {"role": "user", "content": query}
        ],
    )
    return response.choices[0].message.content

# Display the result
print("\nAnswer Based on Documentation:\n")
print(get_response_from_context(user_query))

'''
compliance_data.txt-

General Data Protection Regulation (GDPR) requires the following:
- Obtain user consent before collecting personal data.
- Provide clear privacy notices.
- Allow users to access and delete their data.
- Notify users within 72 hours of a data breach.
- Appoint a Data Protection Officer (DPO) for large-scale processing.

Special rules apply to childrenâ€™s data and international data transfers.


user Question:

What steps should we follow to collect data from EU users?

'''

```
